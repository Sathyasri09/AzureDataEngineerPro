```python
 for name in table_name:
     path ='/mnt/silver/Sales/'+ name
     print(path)
     df=spark.read.format('delta').load(path)

     column_names = df.columns

     for old_col_names in column_names:
         new_col_name="".join(["_" + char if char.isupper() and not old_col_names[i -1].isupper() else char for i, char in enumerate(old_col_names)]).lstrip("_")

         df = df.withColumnRenamed(old_col_names, new_col_name)
     
     output_path = '/mnt/gold/Sales/'+ name + '/'
     df.write.format('delta').mode('overwrite').save(output_path)
```
